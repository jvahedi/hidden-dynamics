```
class TreeNode:
  #This TreeNode is the basic building block of the euqation tree.
  def __init__(self, child = None, sibling = None, nodeType = 'Undefined', value = None, symbol = 'Undefned' ):
    """
    This is the init of TreeNode object. 
    """
    self.child = child
    self.sibling = sibling
    self.nodeType = nodeType
    self.value = value
    self.symbol = symbol
  
  def printTree(self):
    #Print the equation corresponding to the expression tree consider current node
    #as the root in str.
    if self.nodeType == 'Undefined':
      raise Exception('nodeType not specified')
    elif self.nodeType == '+':
      if self.child == None or self.sibling == None:
        raise Exception('Operation + missing child or sibling')
      else:
        return '('+self.child.printTree()+')+('+self.sibling.printTree()+')'

    elif self.nodeType == '-':
      if self.child == None or self.sibling == None:
        raise Exception('Operation - missing child or sibling')
      else:
        return '('+self.child.printTree()+')-('+self.sibling.printTree()+')'
    
    elif self.nodeType == '*':
      if self.child == None or self.sibling == None:
        raise Exception('Operation * missing child or sibling')
      else:
        return '('+self.child.printTree()+')*('+self.sibling.printTree()+')'

    elif self.nodeType == '/':
      if self.child == None or self.sibling == None:
        raise Exception('Operation / missing child or sibling')
      else:
        return '('+self.child.printTree()+')/('+self.sibling.printTree()+')'
      
    elif self.nodeType == 'sin':
      if self.child == None:
        raise Exception('Operation sin missing child')
      else:
        return 'sin('+self.child.printTree()+')'

    elif self.nodeType == 'cos':
      if self.child == None:
        raise Exception('Operation cos missing child')
      else:
        return 'cos('+self.child.printTree()+')'

    elif self.nodeType == 'pow':
      if self.child == None:
        raise Exception('Operation power missing child')
      elif self.value == None:
        raise Exception('Operation power value missing')
      else :
        return '('+self.child.printTree()+')^('+str(self.value)+')'

    elif self.nodeType == 'constant':
      if self.value == None:
        raise Exception('Constant not specified')
      else:
        return str(self.value)

    elif self.nodeType == 'variable':
      if self.symbol == 'Undefined':
        raise Exception('Variable not specified')
      else:
        return self.symbol

    else:
      raise Exception('Unknown tree node type')


  def treeExpr(self):
    #Convert the expression tree consider current node as root in to Sympy Lambda expression 
    if self.nodeType == 'Undefined':
      raise Exception('nodeType not specified')
    elif self.nodeType == '+':
      if self.child == None or self.sibling == None:
        raise Exception('Operation + missing child or sibling')
      else:
        return (self.child.treeExpr()) + (self.sibling.treeExpr())

    elif self.nodeType == '-':
      if self.child == None or self.sibling == None:
        raise Exception('Operation - missing child or sibling')
      else:
        return (self.child.treeExpr()) - (self.sibling.treeExpr())
    
    elif self.nodeType == '*':
      if self.child == None or self.sibling == None:
        raise Exception('Operation * missing child or sibling')
      else:
        return (self.child.treeExpr()) * (self.sibling.treeExpr())

    elif self.nodeType == '/':
      if self.child == None or self.sibling == None:
        raise Exception('Operation / missing child or sibling')
      else:
        return (self.child.treeExpr()) / (self.sibling.treeExpr())
      
    elif self.nodeType == 'sin':
      if self.child == None:
        raise Exception('Operation sin missing child')
      else:
        return sin(self.child.treeExpr())

    elif self.nodeType == 'cos':
      if self.child == None:
        raise Exception('Operation cos missing child')
      else:
        return cos(self.child.treeExpr())

    elif self.nodeType == 'pow':
      if self.child == None:
        raise Exception('Operation power missing child')
      elif self.value == None:
        raise Exception('Operation power value missing')
      else :
        return (self.child.treeExpr())^self.value

    elif self.nodeType == 'constant':
      if self.value == None:
        raise Exception('Constant not specified')
      else:
        return self.value

    elif self.nodeType == 'variable':
      if self.symbol == 'Undefned':
        raise Exception('Variable not specified')
      else:
        variable = symbols(str(self.symbol))
        return variable

    else:
      raise Exception('Unknown tree node type:'+self.nodeType)

  def copyTree(self):
    #copy the tree start from current Node
    newRoot = TreeNode(nodeType = (self.nodeType+'.')[:-1], value = self.value, symbol = (self.symbol+'.')[:-1])
    if self.child != None:
      newRoot.child = self.child.copyTree()

    if self.sibling != None:
      newRoot.sibling = self.sibling.copyTree()
    return newRoot

  def evalTree(self, symbols, values):
    return 1

class TreeOperator:
  def __init__(self, treeRoots = None, maxHight = 5, population = 100, listF = ['+', '-', '*', '/', 'sin'],\
               weightF = None, listT = ['constant', 'x1'], weightT = None, constantRange = 100, fullTRatio = None):
    self.treeRoots = treeRoots
    self.maxHight = maxHight
    self.population = population
    self.listF = listF
    self.listT = listT
    self.constantRange = constantRange
    if weightF == None:
      sizeF = len(listF)
      self.weightF = [1/sizeF] * sizeF
    else:
      self.weightF = weightF
    if weightT == None:
      sizeT = len(listT)
      self.weightT = [1/sizeT] * sizeT
    else:
      self.weightT = weightT
    if fullTRatio == None:
      self.fullTRatio = lambda x: x*0 + 0.5

  def randomFull(self, treeHight = 5):
    root = TreeNode()
    if treeHight >=2:
      nodeType = random.choices(population = self.listF, weights = self.weightF,k = 1)[0]
      root.nodeType = nodeType
      if nodeType == 'sin':
        child = self.randomFull(treeHight - 1)
        root.child = child
      else:
        child = self.randomFull(treeHight - 1)
        sibling = self.randomFull(treeHight - 1)
        root.child = child
        
        while nodeType == '/' and sibling == 0:
          sibling = self.randomFull(treeHight - 1)
        root.sibling = sibling
    else:
      nodeType = random.choices(population = self.listT, weights = self.weightT, k = 1)[0]
      if nodeType == 'constant':
        root.nodeType = nodeType
        root.value = random.random()*self.constantRange
      else:
        root.nodeType = 'variable'
        root.symbol = nodeType
    return root

  def randomGrow(self, treeHight = 5, currentMax = 5):
    r = random.random()
    root = TreeNode()
    if (r < self.fullTRatio(treeHight/currentMax) or treeHight == 1) and (not treeHight == currentMax):
      nodeType = random.choices(population = self.listT, weights = self.weightT, k = 1)[0]
      if nodeType == 'constant':
        root.nodeType = nodeType
        root.value = random.random()*self.constantRange
      else:
        root.nodeType = 'variable'
        root.symbol = nodeType
    else:
      nodeType = random.choices(population = self.listF, weights = self.weightF,k = 1)[0]
      root.nodeType = nodeType
      if nodeType == 'sin':
        child = self.randomGrow(treeHight - 1)
        root.child = child
      else:
        child = self.randomGrow(treeHight - 1)
        sibling = self.randomGrow(treeHight - 1)
        root.child = child
        
        while nodeType == '/' and sibling == 0:
          sibling = self.randomGrow(treeHight - 1)
        root.sibling = sibling
    return root

  def randomHH(self):
    self.treeRoots = []
    kindPopulation = int(self.population /((self.maxHight - 1) * 2))
    for i in range(2, self.maxHight+1):
      for j in range(kindPopulation):
        self.treeRoots.append(self.randomFull(treeHight = i))
      for k in range(kindPopulation):
        self.treeRoots.append(self.randomGrow(treeHight = i, currentMax = i))

  def listOutFT(self, root, listF, result = [[],[]]):
    if root.child != None:
      if root.child.nodeType in listF:
        result[0].append(root.child)
      else:
        result[1].append(root.child)
      self.listOutFT(root.child, listF, result)
    if root.sibling != None:
      if root.sibling.nodeType in listF:
        result[0].append(root.sibling)
      else:
        result[1].append(root.sibling)
      self.listOutFT(root.sibling, listF, result)
    return result

  def crossOver(self, root_1, root_2, FTRatio = 0.9):
    r1 = random.random()
    r2 = random.random()
    r3 = random.random()
    r4 = random.random()
    new_root_1 = root_1.copyTree()
    new_root_2 = root_2.copyTree()
    list1 = self.listOutFT(new_root_1, self.listF)
    list2 = self.listOutFT(new_root_2, self.listF)
    if r1 < FTRatio :
      r2 = math.floor(r2*len(list1[0]))
      selection_1 = list1[0][int(r2)]
    else :
      r2 = math.floor(r2*len(list1[1]))
      selection_1 = list1[1][int(r2)]
    if r3 < FTRatio :
      r4 = math.floor(r4*len(list2[0]))
      selection_2 = list2[0][int(r4)]
    else :
      r4 = math.floor(r4*len(list2[1]))
      selection_2 = list2[1][int(r4)]
    copy_s1 = selection_1.copyTree()
    copy_s2 = selection_2.copyTree()
    selection_1.nodeType = copy_s2.nodeType
    selection_1.value = copy_s2.value
    selection_1.symbol = copy_s2.symbol
    selection_1.child = copy_s2.child
    selection_1.sibling = copy_s2.sibling
    selection_2.nodeType = copy_s1.nodeType
    selection_2.value = copy_s1.value
    selection_2.symbol = copy_s1.symbol
    selection_2.child = copy_s1.child
    selection_2.sibling = copy_s1.sibling
    return new_root_1, new_root_2

  def var_1_L2Norm(root, value, variable, symbol = 'x1'):
    x1 = symbols(symbol)
    expr = root.treeExpr()
    f = lambdify(x1, expr)
    try:
      a = f(variable) - value
      result = np.linalg.norm(a)
    except:
      print('zero divesion')
      result = 1000000000
    
    return result

  def keyFinder(self, element):
    return element[0]

  def nextGen(self, value, variable, norm = var_1_L2Norm):
    rootList = []
    for root in self.treeRoots:
      print(root.printTree())
      rootList.append((norm(root = root, value = value, variable = variable), root))
    rootList.sort(key = self.keyFinder)
    self.treeRoots = []
    replicateSize = int(math.floor(self.population*0.1))
    for i in range(replicateSize):
      self.treeRoots.append(rootList[i][1])
    
    for i in range(int((self.population - replicateSize)/2)):
      root_1 = self.treeRoots[int(math.floor(random.random() * replicateSize))]
      root_2 = self.treeRoots[int(math.floor(random.random() * replicateSize))]
      root_3, root_4 = self.crossOver(root_1, root_2)
      self.treeRoots.append(root_3)
      self.treeRoots.append(root_4)

  


  def printPopulation(self):
    for root in self.treeRoots:
      print(root.printTree())

  def partialDerivative(self, symbol):
    var = symbols(symbol)
    treeExpr = self.treeRoot.treeExpr()
    return Derivative(treeExpr, var).doit()
```
